```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
train_df = pd.read_csv("train.csv")
```

Drop the column and row titled 'label' for both the X and y

```python
X = train_df.drop('label', axis=1)
X = X.to_numpy();
y = train_df['label']
y = y.to_numpy();
```

Feature scaling: this is actually min-max, where min is 0 and max is 255

```python
X = X / 255
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

BELOW IS ACTUAL ML STUFF

SVM

LINEAR KERNEL

```python
X_tune = X_train[:4200]
y_tune = y_train[:4200]
svm_lin_scores = []
svm_lin_c_values = (.001, .003, .01, .1, 1)
for C_ in svm_lin_c_values:
    linear_model = SVC(kernel='linear', C=C_)  # specify model
    svm_lin_scores.append(np.mean(cross_val_score(linear_model, X_tune, y_tune, cv=3, n_jobs=-1)))

print(svm_lin_scores)
```

```python
plt.plot(svm_lin_c_values, svm_lin_scores)
plt.title("Accuracy vs C")
plt.xlabel("C")
plt.ylabel("accuracy")
plt.xscale('log')
plt.show()
```

```python
svm_lin_C_ = svm_lin_c_values[np.argmax(svm_lin_scores)]
print(svm_lin_C_)
# C = .01
```

```python
# linear model
model_linear = SVC(kernel='linear', C =.01)
model_linear.fit(X_train, y_train)
```

```python
y_pred = model_linear.predict(X_test)
```

```python
from sklearn import metrics
acc= metrics.accuracy_score(y_test, y_pred)
print("accuracy:","{:.2f}".format(acc*100),"%")
```

GAUSSIAN KERNEL

Hyperparameter Tuning



```python
from sklearn.model_selection import cross_val_score
X_tune = X_train[:4200]
y_tune = y_train[:4200]
svm_gaus_scores = []
svm_gaus_c_values = (.01, .1, 1, 10, 50)
svm_gaus_gamma_values = (.05, .03, .01, .005, 1/784, .001)
for C_ in svm_gaus_c_values:
    temp = []
    for gamma_ in svm_gaus_gamma_values:
        rbf_model = SVC(kernel='rbf', C=C_, gamma=gamma_)  # specify model
        temp.append(np.mean(cross_val_score(rbf_model, X_tune, y_tune, cv=3, n_jobs=-1)))
    svm_gaus_scores.append(temp)

print(svm_gaus_scores)
```

```python
for i, C_ in enumerate(svm_gaus_scores):
    plt.plot(svm_gaus_gamma_values, C_);
    plt.title("Gamma vs Accuracy when C = " + str(svm_gaus_c_values[i]))
    plt.xlabel("gamma")
    plt.ylabel("accuracy")
    plt.show()

svm_gaus_c_scores = [np.max(i) for i in svm_gaus_scores]
plt.plot(svm_gaus_c_values, svm_gaus_c_scores)
plt.title("C vs Accuracy")
plt.xscale("log")
```

Now that our hyperparameters have been found, run on all data!

```python
svm_gaus_C_ = svm_gaus_c_values[np.argmax(svm_gaus_c_scores)]
print(svm_gaus_C_)
svm_gaus_gamma_ = svm_gaus_gamma_values[np.argmax(svm_gaus_scores[np.argmax(svm_gaus_c_scores)])]
print(svm_gaus_gamma_)
# C = 10, gamma = .03
```

```python
rbf_model = SVC(kernel='rbf', C=svm_gaus_C_, gamma=svm_gaus_gamma_)  # specify model
rbf_model.fit(X_train, y_train)
```

```python
# predict the value
y_rbf_pred = rbf_model.predict(X_test)
```

```python
# Accuracy
from sklearn import metrics
acc_rbf= metrics.accuracy_score(y_test, y_rbf_pred)
print("accuracy:","{:.2f}".format(acc_rbf*100),"%")
```
