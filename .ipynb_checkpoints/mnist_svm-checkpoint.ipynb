```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
train_df = pd.read_csv("train.csv")
```

Drop the column and row titled 'label' for both the X and y

```python
X = train_df.drop('label', axis=1)
X = X.to_numpy();
y = train_df['label']
y = y.to_numpy();
# X = X[:4667]
# y = y[:4667]
```

Feature scaling: this is actually min-max, where min is 0 and max is 255

```python
X = X / 255
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

Linear Kernel
Takes about a minute to run.

```python
from sklearn.svm import SVC
```

```python
# linear model
model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)
```

```python
y_pred = model_linear.predict(X_test)
```

```python
from sklearn import metrics
acc= metrics.accuracy_score(y_test, y_pred)
print("accuracy:","{:.2f}".format(acc*100),"%")
```

Hyperparameter Tuning TESTING

With 4,200 data samples for training/CV and 467 for testing (my hyperparameters) -> 95.93%


4,200 (default) -> 94.86%


With 42,000 data samples, split 70:30 (my hyperparameters) -> 98.12


With 42,000 data samples, split 70:30 -> 97.38%



```python
from sklearn.model_selection import cross_val_score
X_tune = X_train[:4200]
y_tune = y_train[:4200]
scores = []
c_values = (.01, .1, 1, 10, 50)
gamma_values = (.05, .03, .01, .005, 1/784, .001)
for C_ in c_values:
    temp = []
    for gamma_ in gamma_values:
        rbf_model = SVC(kernel='rbf', C=C_, gamma=gamma_)  # specify model
        temp.append(np.mean(cross_val_score(rbf_model, X_tune, y_tune, cv=3, n_jobs=-1)))
    scores.append(temp)

print(scores)
```

```python
for i, C_ in enumerate(scores):
    plt.plot(gamma_values, C_);
    plt.title("C = " + str(c_values[i]))
    plt.xlabel("gamma")
    plt.ylabel("accuracy")
    plt.show()

c_scores = [np.max(i) for i in scores]
plt.plot(c_values, c_scores)
```

Now that our hyperparameters have been found, run on all data!

```python
C_ = c_values[np.argmax(c_scores)]
print(C_)
gamma_ = gamma_values[np.argmax(scores[np.argmax(c_scores)])]
print(gamma_)
```

```python
rbf_model = SVC(kernel='rbf', C=C_, gamma=gamma_)  # specify model
rbf_model.fit(X_train, y_train)
```

```python
# predict the value
y_rbf_pred = rbf_model.predict(X_test)
```

```python
# Accuracy
from sklearn import metrics
acc_rbf= metrics.accuracy_score(y_test, y_rbf_pred)
print("accuracy:","{:.2f}".format(acc_rbf*100),"%")
```
